```python
import pandas as pd
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.info()
```

```python
filtered_data = avocado.query('year == 2015 and type == "conventional" and region =="Albany" ')
filtered_data = filtered_data [['Date', 'Total Bags', 'Small Bags']]
filtered_data.head()
```

```python
melted_df = avocado.melt(id_vars=['Date'], value_vars=['Total Bags','Small Bags'], var_name='Bags',value_name='Count')
melted_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('mode.use_inf_as_na', True)
sns.lineplot (data= melted_df,x='Date',y='Count',hue='Bags')
plt.show()
```

```python
bin_labels = ['Poor', 'Modest', 'Good', 'Excellent']
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q=4, labels=bin_labels)
avocado.head()
```

```python
avocado['Sales Volume'] = avocado['Sales Volume'].astype('category')
sns.countplot(data=avocado, x='year', hue='Sales Volume')
```

```python
melted_df = pd.melt(avocado, id_vars = 'year', value_vars=['Small Bags', 'Large Bags','XLarge Bags'], var_name='Bags', value_name='Count')
melted_df
```

```python

sns.relplot(data=melted_df,kind= 'line',x='year', y= 'Count', hue= 'Bags')
```
